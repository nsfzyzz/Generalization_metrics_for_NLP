# Try to see how the validation loss goes

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from re import search
```

```python
def analyze_one_file(file_name):
    
    val_loss = []
    train_loss = []

    with open(file_name, 'r') as f:
        lines = f.readlines()
        for line in lines:
            x = line.split(" ")
            if "Val" in x and "loss:" in x:
                val_loss.append(float(x[-2]))
            if "Average" in x and "training" in x:
                train_loss.append(float(x[-1]))
    
    return train_loss, val_loss
```

```python
def get_ww_result(plot_result, ww_result, keys):

    print(ww_result)
    with open(ww_result, 'rb') as f:
        dict0 = pickle.load(f)
        for key in keys:
            if key!='rand_num_spikes':
                plot_result[key].append(dict0['summary'][key])
            else:
                spike_values = [dict0['details']['rand_num_spikes'][key] for key in dict0['details']['rand_num_spikes'].keys()]
                plot_result[key].append(sum(spike_values))
```

```python
def plot_once(ax, title, results):
    ax.set_title(title, fontsize=14)
    for key in results.keys():
        num_epochs = len(results[key])
        if search('Training', title):
            ax.plot(np.arange(0.5, 0.5+num_epochs), results[key], label=key)
        else:
            ax.plot(np.arange(1, 1+num_epochs), results[key], label=key)
    if search('Alpha', title):
        ax.set_ylim([2, 6])
    ax.set_xlabel('Training epochs', fontsize=14)
    if search('Training', title) or search('Validation', title):
        ax.set_ylabel('Cross entropy loss', fontsize=14)
    elif search('Alpha', title):
        ax.set_ylabel('alpha', fontsize=14)
    elif search('rand_num_spikes', title):
        ax.set_ylabel('Number of spikes', fontsize=14)
    
    ax.legend(fontsize=14)
```

```python
def add_keys(results):
    keys = ['alpha', 'rand_num_spikes']
    for key in keys:
        results[key] = []
```

```python
fig, axs = plt.subplots(1,4, figsize=(20,5))

train_loss_clean, val_loss_clean = analyze_one_file('logs/clean.log')
train_loss_sharpness, val_loss_sharp = analyze_one_file('logs/sharpness.log')


plot_once(axs[0], title='Training loss', results={'SVDSharpness': train_loss_sharpness, 
                                                 'No SVDSharpness': train_loss_clean})
plot_once(axs[1], title='Validation loss', results={'SVDSharpness': val_loss_sharp, 
                                                 'No SVDSharpness': val_loss_clean})
results_clean = {}
add_keys(results_clean)
for epoch in range(10):
    ww_result = f'../checkpoint/BERT_finetune/clean/metrics/epoch_{epoch}/results.pkl'
    get_ww_result(results_clean, ww_result, keys)
    
results_sharp = {}
add_keys(results_sharp)
for epoch in range(10):
    ww_result = f'../checkpoint/BERT_finetune/sharp/metrics/epoch_{epoch}/results.pkl'
    get_ww_result(results_sharp, ww_result, keys)

plot_once(axs[2], title='Alpha', results={'SVDSharpness': results_sharp['alpha'], 
                                            'No SVDSharpness': results_clean['alpha']})
plot_once(axs[3], title='rand_num_spikes', results={'SVDSharpness': results_sharp['rand_num_spikes'], 
                                            'No SVDSharpness': results_clean['rand_num_spikes']})
```

# SVDSmoothing

```python
val_loss_02 = []
val_loss_04 = []
val_loss_07 = []
val_loss_sanity_check = []

_, val_loss = analyze_one_file(f'logs/clean.log')

for epoch in range(10):
    _, val_loss_one_02 = analyze_one_file(f'logs/SVDSmoothing_epoch_{epoch}.log')
    _, val_loss_one_04 = analyze_one_file(f'logs/SVDSmoothing_0.4_epoch_{epoch}.log')
    _, val_loss_one_07 = analyze_one_file(f'logs/SVDSmoothing_0.7_epoch_{epoch}.log')
    _, val_loss_sanity_check_one = analyze_one_file(f'logs/SVDSmoothing_tail_num_epoch_{epoch}.log')
    val_loss_02.append(val_loss_one_02)
    val_loss_04.append(val_loss_one_04)
    val_loss_07.append(val_loss_one_07)
    val_loss_sanity_check.append(val_loss_sanity_check_one)
```

```python
val_loss_02 = [x[0] for x in val_loss_02]
val_loss_04 = [x[0] for x in val_loss_04]
val_loss_07 = [x[0] for x in val_loss_07]
val_loss_sanity_check = [x[0] for x in val_loss_sanity_check]
```

```python
plt.figure()
plt.plot(np.arange(1,11), val_loss_02, label='Smoothing percent=0.2')
plt.plot(np.arange(1,11), val_loss_04, label='Smoothing percent=0.4')
plt.plot(np.arange(1,11), val_loss_07, label='Smoothing percent=0.7')
plt.plot(np.arange(1,11), val_loss, label='val loss')
plt.plot(np.arange(1,11), val_loss_sanity_check, label='SVDSmoothing (tail num)')
plt.plot(np.arange(0.5,10.5), train_loss_clean, label='Training loss')
plt.legend(loc='center left')
plt.ylabel('Loss', fontsize=14)
plt.xlabel('Epochs', fontsize=14)
plt.title('SVDSmoothing, BERT SST-2', fontsize=14)
plt.ylim([-0.02, 0.9])
```
